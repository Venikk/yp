```python
datasets_path = ''
#datasets_path = 'C:/Users/Venik/OneDrive/–î–æ–∫—É–º–µ–Ω—Ç—ã/Yandex_Praktikum/_Final'
```

```python
contract = pd.read_csv(datasets_path + '/datasets/final_provider/contract.csv')
contract.sample(5)
```

```python
contract.info()
```

```python
# –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–æ–∫ –∏ –¥–∞–º–º–∏-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ
contract.set_axis(['customer_id','begin_date','end_date',
                   'contract_type', 'paperless_billing','payment_method',
                  'monthly_charges','total_charges'],
                  axis = 'columns',inplace = True)

contract_with_dummies = pd.get_dummies(contract, 
                                       columns=['contract_type','paperless_billing','payment_method'], 
                                       drop_first=True)

contract_with_dummies.sample(5)
```

```python
contract_with_dummies.set_axis(['customer_id','begin_date','end_date', 'monthly_charges','total_charges',
                                'contract_type_one_year', 'contract_type_two_year', 'paperless_billing',
                                'payment_method_credt_card', 'payment_method_electronic_check',
                                'payment_method_mailed_check'],
                                axis = 'columns',inplace = True)
contract_with_dummies.info()
```

```python
# –§–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –∏ —É–¥–∞–ª–µ–Ω–∏–µ –ª–∏—à–Ω–∏—Ö –∫–æ–ª–æ–Ω–æ–∫
def otval(row):
    if row['end_date'] == 'No':
        return 0
    else:
        return 1
    
contract_with_dummies['otval'] = contract_with_dummies.apply(otval, axis=1)
contract_with_dummies['begin_date_d'] = pd.to_datetime(contract_with_dummies['begin_date'], format='%Y-%m-%d')
contract_with_dummies['end_date'].replace('No', '2020-02-01 00:00:00',inplace=True);
contract_with_dummies['end_date_d'] = pd.to_datetime(contract_with_dummies['end_date'], format='%Y.%m.%d %H:%M:%S')

contract_with_dummies['life_time'] = (contract_with_dummies['end_date_d'] - 
                                      contract_with_dummies['begin_date_d']).dt.total_seconds() / 86400

contract_with_dummies['life_time'] = contract_with_dummies['life_time'].astype(int)

contract_with_dummies.drop(['begin_date','end_date','total_charges','begin_date_d','end_date_d'], axis=1, inplace = True)
contract_with_dummies.sample(5)
```

### –ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –æ–¥–∏–Ω —Ñ–∞–π–ª —Å–æ –≤—Å–µ–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –∏ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –µ–≥–æ.

2. –õ–µ–≤—ã–º —Å–æ–µ–¥–∏–Ω–µ–Ω–∏–µ–º –¥–æ–±–∞–≤–∏–º –¥–∞–Ω–Ω—ã–µ –∏–∑ —Ç–∞–±–ª–∏—Ü —Å –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –∏–Ω—Ç–µ–Ω–µ—Ç—É –∏ —Ç–µ–ª–µ—Ñ–æ–Ω–∏–∏.
3. –£–¥–∞–ª–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –¥–∞–Ω–Ω—ã—Ö –ø–æ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç—É –∏ —Ç–µ–ª–µ—Ñ–æ–Ω–∏–∏. –ï—Å–ª–∏ –∫–ª–∏–µ–Ω—Ç —É –∫–ª–∏–µ–Ω—Ç–∞ –Ω–µ—Ç –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è —ç—Ç–æ–≥–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è —É—Å–ª—É–≥, —Ç–æ –≤—Å–µ –∑–∞–≤–∏—Å–∏–º—ã–µ –ø–æ–ª—è —Å —É—Å–ª—É–≥–∞–º–∏ —ç—Ç–æ–≥–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è —É—Å—Ç–∞–Ω–æ–≤–∏–º –≤ -1. 

```python
# –°–∫–ª–µ–∏–≤–∞–Ω–∏–µ —Ç–∞–±–ª–∏—Ü –¥–∞–Ω–Ω—ã—Ö –æ –∫–æ–Ω—Ç—Ä–∞–∫—Ç–∞—Ö –∏ —Ç–∞–±–ª–∏—Ü —Å –ø–µ—Ä—Å–æ–Ω–∞–ª—å–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
personal_contract = personal_with_dummies.merge(contract_with_dummies, how = 'left', 
                                                left_on = 'customer_id', right_on = 'customer_id')
personal_contract.info()
```

```python
personal_contract.sample(5)
```

```python
# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –æ —É—Å–ª—É–≥–∞—Ö —Ç–µ–ª–µ—Ñ–æ–Ω–∏–∏
personal_contract_phone = personal_contract.merge(phone_with_dummies, how = 'left', 
                                                  left_on = 'customer_id', right_on = 'customer_id')
personal_contract_phone.info()
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã 
def df_hystogram(series,
                 series_name,
                 description_name,
                 diagram_color,
                 bins_number,
                 compare_data):
    
    n, bins, patches = plt.hist(series, 
                                   bins_number, 
                                   density=True, 
                                   facecolor=diagram_color, 
                                   label = series_name, 
                                   alpha=0.3)
    
    series_description = series.describe()

    compare_data = compare_data.append(pd.DataFrame([[description_name, 
                                                      series_name, 
                                                      series_description[1]]], 
                                                      columns = ['x', 
                                                                'hue',
                                                                'y']),
                                                      ignore_index=True)

    y = scs.norm.pdf(bins, series_description[1], series_description[2])

    plt.plot(bins, y, '--'+diagram_color)

    return series_description, compare_data

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—è —Ç—Ä–µ—Ö –¥–∏–∞–≥—Ä–∞–º —Ä—è–¥–æ–º
def df_two_histograms(series1, 
                      series2, 
                      series_name1, 
                      series_name2, 
                      description_name, 
                      bins_number,
                      xlim_min,
                      xlim_max,
                      ylim_max,
                      compare_data):

    plt.figure(0)

    series1_description, compare_data = df_hystogram(series1,
                                                   series_name1,
                                                   description_name,
                                                   'g',
                                                   bins_number,
                                                   compare_data)

    series2_description, compare_data = df_hystogram(series2,
                                                   series_name2,
                                                   description_name,
                                                   'b',
                                                   bins_number,
                                                   compare_data)

    
    plt.xlabel('–°—Ä–µ–¥–Ω–µ–º–µ—Å—è—á–Ω—ã–π —á–µ–∫')
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞ –≤ –≤—ã–±–æ—Ä–∫–µ')

    plt.gca().set(title=description_name)
    
    if xlim_max != 0:
        plt.xlim(right=xlim_max) 
        
    if xlim_min != 0:
        plt.xlim(left=xlim_min)

    if ylim_max != 0:
        plt.ylim(top=ylim_max) 
        
    
    plt.legend()
    plt.grid()
    plt.plot()
    
    return series1_description, series2_description, compare_data
```

```python
# –°–æ–∑–¥–∞–¥–∞–∏–º –º–∞—Å—Å–∏–≤ —Å—Ç—Ä–æ–∫ –¥–ª—è –∑–∞–ø–∏—Å–∏ —Å—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
rows_list = []

# –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è —Å—Ç—Ä–æ–∫–∏ –≤ –º–∞—Å—Å–∏–≤
def df_statistic_append(dataframe,
                        column_name,
                        sample_column_name,
                        sample_name,
                        indicator_name):
    
    local_series = dataframe[column_name]
    
    series_description = local_series.describe()

    local_dict = {sample_column_name: sample_name,
                  'indicator': indicator_name, 
                  'mean': series_description[1],
                  'median': series_description[5],
                  'min': series_description[3],
                  'low_kw': series_description[4],
                  'high_kw': series_description[6],
                  'max': series_description[7],
                  'std': series_description[2]
                 }
    rows_list.append(local_dict)

# –î–æ–±–∞–≤–∏–º —Å—Ç—Ä–æ–∫–∏ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
df_statistic_append(data_final,
                    'monthly_charges',
                    'dataset',
                    '–í—Å–µ –∫–ª–∏–µ–Ω—Ç—ã',
                    '–°—Ä–µ–¥–Ω–µ–º–µ—Å—è—á–Ω—ã–π —á–µ–∫')

df_statistic_append(data_otval,
                    'monthly_charges',
                    'dataset',
                    '–û—Ç–≤–∞–ª–∏–≤—à–∏–µ—Å—è –∫–ª–∏–µ–Ω—Ç—ã',
                    '–°—Ä–µ–¥–Ω–µ–º–µ—Å—è—á–Ω—ã–π —á–µ–∫')

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –º–∞—Å—Å–∏–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
statistic_monthly_charge = pd.DataFrame(rows_list)

statistic_monthly_charge
```

1. –†–∞–∑–æ–±—ä–µ–º –≤—ã–±–æ—Ä–∫—É –Ω–∞ –æ–±—É—á–∞—é—â—É—é, –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é.
2. –ù–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –æ–±—É—á–∏–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –º–æ–¥–µ–ª–µ–π —Å –±–∞–∑–æ (–ø–æ–∫–∞ –Ω–µ –æ–ø—Ä–µ–¥–µ–ª–∏–ª—Å—è –∫–∞–∫–∏—Ö) –∏ –≤—ã–±–µ—Ä–µ–º –ª—É—á—à—É—é.
3. –î–ª—è —ç—Ç–æ–π –º–æ–¥–µ–ª–∏ –ø–æ–¥–±–µ—Ä–µ–º –ª—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –ª—É—á—à–∏–π AUC-ROC –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ.
4. –í—ã—á–∏—Å–ª–∏–º AUC-ROC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ.
5. –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –∑–Ω–∞—á–∏–º–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞.

```python
features_train_valid, features_test, target_train_valid, target_test = train_test_split(
    features, target, test_size=0.20, random_state=12345)

features_train, features_valid, target_train, target_valid = train_test_split(
    features_train_valid, target_train_valid, test_size=0.25, random_state=12345)

print(features_train.shape, end = ' '); print(target_train.shape)
print(features_valid.shape, end = ' '); print(target_valid.shape)
print(features_test.shape, end = ' '); print(target_test.shape)
```

```python
# –°–æ–∑–¥–∞–¥–∏–º –¥–∞—Ç–∞—Å–µ—Ç –¥–ª—è –∑–∞–ø–∏—Å–∏ —Å—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π
# model - –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏
# model_param_name - –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è –º–æ–¥–µ–ª–∏
# model_param_value - –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è –º–æ–¥–µ–ª–∏
# accuracy - –∑–Ω–∞—á–µ–Ω–∏–µ accuracy
# auc_roc - –∑–Ω–∞—á–µ–Ω–∏–µ AUC-ROC

compare = {'model' : [],
           'param1_name' : [],
           'param1_value' : [],
           'param2_name' : [],
           'param2_value' : [],
           'accuracy' : [],
           'auc_roc' : []
           }

compare_data = pd.DataFrame(compare)
```

```python
# –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å—á–µ—Ç–∞ auc_roc
# model_income - –º–æ–¥–µ–ª—å, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–π —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏
# features_income - –º–∞—Å—Å–∏–≤ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞
# target_income - –º–∞—Å—Å–∏–≤ —Ü–µ–ª–µ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞

def count_auc_roc(model_income, 
                  features_income,
                  target_income):
    
    probabilities_valid = model_income.predict_proba(features_income)
    probabilities_one_valid = probabilities_valid[:, 1]

    return roc_auc_score(target_income, probabilities_one_valid)

# –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å—á–µ—Ç–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–∏ –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –∏—Ö –≤ —Å—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç
# target_income - —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫
# predicted_income - –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫
# compare_data_income - –¥–∞—Ç–∞—Å–µ—Ç –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è —Å—Ç—Ä–æ–∫–∏
# model_name - –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏
# model_param_name - –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è –º–æ–¥–µ–ª–∏
# model_param_value - –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è –º–æ–¥–µ–ª–∏
# f1_income - –≤—Ö–æ–¥—è—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ f1-–º–µ—Ä—ã

def count_model_params(model_income,
                       features_train_income,
                       target_train_income, 
                       features_valid_income,
                       target_valid_income, 
                       compare_data_income,
                       model_name,
                       param1_name,
                       param1_value,
                       param2_name,
                       param2_value,
                       auc_roc_income):

    model.fit(features_train_income, target_train_income)

    predicted_valid = model.predict(features_valid_income)

    auc_roc_counted = count_auc_roc(model, 
                                    features_valid_income,
                                    target_valid_income)        
    accuracy_counted = accuracy_score(target_valid_income, predicted_valid)

    if (auc_roc_counted > auc_roc_income):
        compare_data_outcome = compare_data_income.append(
            pd.DataFrame([[model_name, 
                           param1_name,
                           param1_value,
                           param2_name,
                           param2_value,
                           accuracy_counted,
                           auc_roc_counted]], 
                           columns = ['model', 
                                       'param1_name',
                                       'param1_value',
                                       'param2_name',
                                       'param2_value',
                                       'accuracy',
                                       'auc_roc']),
                            ignore_index=True)
        
        return compare_data_outcome, auc_roc_counted
    
    else:
        
        return compare_data_income, auc_roc_income
        
```

```python
# –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

# –£—Å—Ç–∞–Ω–æ–≤–∏–º —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ AUC-ROC –≤ 0.
# –í —Å—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç –±—É–¥–µ–º –¥–æ–±–∞–≤–ª—è—Ç—å —Å—Ç—Ä–æ–∫—É —Ç–æ–ª—å–∫–æ –≤ —Å–ª—É—á–∞–µ,
# –µ—Å–ª–∏ —É –º–æ–¥–µ–ª–∏ —Å–æ —Å–ª–µ–¥—É—é—â–∏–º –Ω–∞–±–æ—Ä–æ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ AUC-ROC –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –±—É–¥–µ—Ç –≤—ã—à–µ –ø—Ä–µ–¥—ã–¥—É—â–µ–π 
auc_roc_current = 0

model = LogisticRegression(random_state=12345)


compare_data, auc_roc_current = count_model_params(model,
                                                   features_train,
                                                   target_train, 
                                                   features_valid,
                                                   target_valid, 
                                                   compare_data,
                                                   '–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è',
                                                   '',
                                                   0,
                                                   '',
                                                   0,
                                                   auc_roc_current)

compare_data
```

```python
# –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

# –£—Å—Ç–∞–Ω–æ–≤–∏–º —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ AUC-ROC –≤ 0.
# –í —Å—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç –±—É–¥–µ–º –¥–æ–±–∞–≤–ª—è—Ç—å —Å—Ç—Ä–æ–∫—É —Ç–æ–ª—å–∫–æ –≤ —Å–ª—É—á–∞–µ,
# –µ—Å–ª–∏ —É –º–æ–¥–µ–ª–∏ —Å–æ —Å–ª–µ–¥—É—é—â–∏–º –Ω–∞–±–æ—Ä–æ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ AUC-ROC –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –±—É–¥–µ—Ç –≤—ã—à–µ –ø—Ä–µ–¥—ã–¥—É—â–µ–π 
auc_roc_current = 0


for estimators in range(110, 200, 5):

    model = RandomForestClassifier(random_state=12345, 
                                   n_estimators=estimators)

    compare_data, auc_roc_current = count_model_params(model,
                                                       features_train,
                                                       target_train, 
                                                       features_valid,
                                                       target_valid, 
                                                       compare_data,
                                                       '–°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å',
                                                       'depth',
                                                       0,
                                                       'n_estimators',
                                                       estimators,
                                                       auc_roc_current)

compare_data
```

```python
# LightGBM

# –£—Å—Ç–∞–Ω–æ–≤–∏–º —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ AUC-ROC –≤ 0.
# –í —Å—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç –±—É–¥–µ–º –¥–æ–±–∞–≤–ª—è—Ç—å —Å—Ç—Ä–æ–∫—É —Ç–æ–ª—å–∫–æ –≤ —Å–ª—É—á–∞–µ,
# –µ—Å–ª–∏ —É –º–æ–¥–µ–ª–∏ —Å–æ —Å–ª–µ–¥—É—é—â–∏–º –Ω–∞–±–æ—Ä–æ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ AUC-ROC –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –±—É–¥–µ—Ç –≤—ã—à–µ –ø—Ä–µ–¥—ã–¥—É—â–µ–π 
auc_roc_current = 0

start_n_estimators = 150
finish_n_estimators = 200
step_n_estimators = 10

start_num_leaves = 25
finish_num_leaves = 36
step_num_leaves = 2

# –ü–æ–¥–±–µ—Ä–µ–º –ª—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
for current_num_leaves in range(start_num_leaves, 
                                        finish_num_leaves, 
                                        step_num_leaves):
    for current_n_estimators in range(start_n_estimators, 
                                                finish_n_estimators, 
                                                step_n_estimators):
                    
        model = LGBMClassifier(random_state=12345, 
                               num_leaves = current_num_leaves,
                               n_estimators = current_n_estimators)    

        compare_data, auc_roc_current = count_model_params(model,
                                                           features_train,
                                                           target_train, 
                                                           features_valid,
                                                           target_valid, 
                                                           compare_data,
                                                           'LightGBM',
                                                           #'max_depth',
                                                           #current_max_depth,
                                                           'num_leaves',
                                                           current_num_leaves,
                                                           'n_estimators',
                                                           current_n_estimators,
                                                           auc_roc_current)

compare_data
```

```python
model = LGBMClassifier(random_state=12345, 
                        max_depth = -1,
                        num_leaves = 27,
                        n_estimators = 190)    

model.fit(features_train, target_train)

predicted_test = model.predict(features_test)
```

```python
probabilities_test = model.predict_proba(features_test)
probabilities_one_test = probabilities_test[:, 1]

fpr, tpr, thresholds = roc_curve(target_test, probabilities_one_test)

plt.figure()

# ROC-–∫—Ä–∏–≤–∞—è –Ω–∞—à–µ–π –º–æ–¥–µ–ª–∏
plt.plot(fpr, tpr, linestyle='-')

# ROC-–∫—Ä–∏–≤–∞—è —Å–ª—É—á–∞–π–Ω–æ–π –º–æ–¥–µ–ª–∏
plt.plot([0, 1], [0, 1], linestyle='--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-–∫—Ä–∏–≤–∞—è –≤—ã–±—Ä–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏')

plt.show()
```

```python
auc_roc_counted = count_auc_roc(model, 
                                features_test,
                                target_test)      

print('AUC-ROC: {:.4f}'.format(auc_roc_counted))
```

```python
accuracy_counted = accuracy_score(target_test, predicted_test)
print('Accuracy: {:.4f}'.format(accuracy_counted))
```
